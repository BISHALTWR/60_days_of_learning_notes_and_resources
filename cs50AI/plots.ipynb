```python
# Step function
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 1000)
y = np.heaviside(x, 0.5)

plt.figure(figsize=(6, 4))
plt.plot(x, y, label='Step function')
plt.title('Step function')
plt.yticks([0,1])
plt.xticks([])
plt.xlabel('w.x')
plt.ylabel('Output y')
plt.grid(True)
plt.legend()

# Add mathematical expression
plt.text(-5, 0.8, r'$g(x) = 1 \; if \; x \geq k \; else \; 0$', fontsize=12)

plt.savefig('step_function.png')
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 1000)
y = 1 / (1 + np.exp(-x))

plt.figure(figsize=(6, 4))
plt.plot(x, y, label='Logistic Sigmoid function')
plt.title('Logistic Sigmoid function')
plt.yticks([0, 1])  # Show 0, 0.5 and 1 on y-axis
plt.xticks([])  # Remove x-axis labels
plt.ylabel('Output y')
plt.xlabel('w.x')
plt.grid(True)
plt.legend()

# Add mathematical expression
plt.text(-5, 0.8, r'$g(x) = \frac{e^x}{e^x + 1}$', fontsize=12)

plt.savefig('logistic_sigmoid_function.png')
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 1000)
y = np.maximum(0, x)

plt.figure(figsize=(6, 4))
plt.plot(x, y, label='ReLU function')
plt.title('ReLU function')
plt.yticks([])  # Show 0 and max on y-axis
plt.xticks([])  # Remove x-axis labels
plt.ylabel('Output y')
plt.grid(True)
plt.legend()

# Add mathematical expression
plt.text(-5, max(y) / 2, r'$g(x) = max(0, x)$', fontsize=12)

plt.savefig('relu_function.png')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a simple binary classification problem
np.random.seed(0)
n = 100  # Number of students
X = np.random.normal(5, 2, (n, 2))  # Increase the standard deviation to 2
X = np.abs(X)  # Make all values positive
y = np.where(X[:, 0] + X[:, 1] > 10, 1, 0)  # Adjust the decision boundary

# Plot the data points
plt.figure(figsize=(8, 6))
colors = np.where(y == 1, 'g', 'r')  # Use green for pass and red for fail
plt.scatter(X[:, 0], X[:, 1], c=colors, edgecolor='k', s=50)

# Draw a decision boundary
x_values = np.linspace(0, 10, 100)
y_values = 10 - x_values
plt.plot(x_values, y_values, 'b--', label='Decision boundary')  # Use blue for the decision boundary

plt.xlim([0, 10])
plt.ylim([0, 10])
plt.title('Pass/Fail Decision Boundary')
plt.xlabel('Hours Studied')
plt.ylabel('Hours of Sleep')
plt.legend()
plt.grid(True)
plt.savefig('perceptron.png')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Create a simple binary classification problem
np.random.seed(0)
n = 100  # Number of students
X = np.random.normal(5, 2, (n, 2))  # Increase the standard deviation to 2
X = np.abs(X)  # Make all values positive
noise = np.random.normal(0, 1, (n, 2))  # Add some noise to the data
X += noise
y = np.where(X[:, 0]**2 + X[:, 1]**2 > 25, 1, 0)  # Adjust the decision boundary

# Plot the data points
plt.figure(figsize=(8, 6))
cmap = ListedColormap(['r', 'g'])  # Use red for fail and green for pass
scatter = plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap, edgecolor='k', s=50)
# plt.colorbar(scatter, ticks=[0,1], label='Pass/Fail')

# Draw a decision boundary
theta = np.linspace(0, 2*np.pi, 100)
x_values = 5 * np.cos(theta)
y_values = 5 * np.sin(theta)
plt.plot(x_values, y_values, 'b--', label='Decision boundary')  # Use blue for the decision boundary

plt.xlim([0, 10])
plt.ylim([0, 10])
plt.title('Pass/Fail Decision Boundary', fontsize=16)
plt.xlabel('Hours Studied', fontsize=14)
plt.ylabel('Hours of Sleep', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```
